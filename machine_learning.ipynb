```python
matches['data'] = pd.to_datetime(matches['data'])
```

```python
matches["local_index"] = matches['local'].astype('category').cat.codes
matches["oponente_index"] = matches['oponente'].astype('category').cat.codes
matches['hora'] = matches['horário'].str.replace(":.+", "", regex=True).astype('int')
matches['dia_index'] = matches['data'].dt.dayofweek

# if matches['resultado'] == 'V':
#     matches['alvo'] = 2
# elif matches['resultado'] == 'D':
#     matches['alvo'] = 1
# else:
#     matches['alvo'] = 0

matches['alvo'] = (matches['resultado'] == 'V').astype('int')
```

```python
matches
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1)

train = matches[matches['data'] < '2024-01-01']
test = matches[matches['data'] > '2024-01-01']

predictors = ['local_index', 'oponente_index', 'hora', 'dia_index']

rf.fit(train[predictors], train['alvo'])

preds = rf.predict(test[predictors])
```

```python
from sklearn.metrics import accuracy_score

acc = accuracy_score(test['alvo'], preds)
```

```python
acc
```

```python
combined = pd.DataFrame(dict(actual=test['alvo'], prediction=preds))
pd.crosstab(index=combined['actual'], columns=combined['prediction'])
```

```python
from sklearn.metrics import precision_score

precision = precision_score(test['alvo'], preds)
```

```python
grouped_matches = matches.groupby('time')
group = grouped_matches.get_group('Corinthians')
group
```

```python
def rolling_averages(group, cols, new_cols):
    group = group.sort_values('data')
    rolling_stats = group[cols].rolling(3, closed="left").mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols)
    
    return group
```

```python
cols = ['gp', 'gc', 'tc', 'cag', 'dist', 'fk', 'pb', 'pt']
new_cols = [f"{c}_rolling" for c in cols]
```

```python
rolling_averages(group, cols, new_cols)
```

```python
matches_rolling = matches.groupby('time').apply(lambda x: rolling_averages(x, cols, new_cols))
```

```python
matches_rolling
```

```python
matches_rolling = matches_rolling.droplevel('time')
matches_rolling.index = range(matches_rolling.shape[0])
```

```python
matches_rolling
```

```python
def make_predictions(data, predictors):
    train = data[data['data'] < '2024-01-01']
    test = data[data['data'] > '2024-01-01']

    rf.fit(train[predictors], train['alvo'])
    preds = rf.predict(test[predictors])

    combined = pd.DataFrame(dict(actual=test['alvo'], prediction=preds))
    precision = precision_score(test['alvo'], preds)

    return combined, precision
```

```python
combined, precision = make_predictions(matches_rolling, predictors + new_cols)
```
